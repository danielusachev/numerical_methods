Реализуем метод Рунге-Кутты четвертого порядка точности

```python
import numpy as np
def solve_system(f,g, x0, y0, z0, x1, n):
    vx = [0] * (n + 1)
    vy = [0] * (n + 1)
    vz = [0] * (n + 1)
    h = (x1 - x0) / float(n)
    vx[0] = x = x0
    vy[0] = y = y0
    vz[0] = z = z0
    for i in range(1, n + 1):
        k1 = h * f(x, y, z)
        l1 = h * g(x, y, z)

        k2 = h * f(x + 0.5 * h, y + 0.5 * k1, z + 0.5 * l1)
        l2 = h * g(x + 0.5 * h, y + 0.5 * k1, z + 0.5 * l1)

        k3 = h * f(x + 0.5 * h, y + 0.5 * k2, z + 0.5 * l2)
        l3 = h * g(x + 0.5 * h, y + 0.5 * k2, z + 0.5 * l2)

        k4 = h * f(x + h, y + k3, z + l3)
        l4 = h * g(x + h, y + k3, z + l3)

        k = 1/6 * (k1 + 2 * k2 + 2 * k3 + k4)
        l = 1/6 * (l1 + 2 * l2 + 2 * l3 + l4)

        vx[i] = x = x0 + i * h
        vy[i] = y = y + k
        vz[i] = z = z + l
    return vx, vy, vz
```

Рассмотри следующую задачу Коши

$
\begin{cases} 
\frac{dy}{dx} = -2y + 4z \\ 
\frac{dz}{dx} = -y + 3z \\ 
y(0) = 3\\
z(0) = 0
\end{cases}
$

Она имеет решение

$
\begin{cases} 
y(x) = 4e^{-x} - e^{2x} \\ 
z(x) = e^{-x} - e^{2x}
\end{cases}
$

```python

def f(x,y,z):
    return -2*y+4*z
def g(x,y,z):
    return -y+3*z
def solution(x,y,z):
    y = 4*np.exp(-x) - np.exp(2*x)
    z = np.exp(-x) - np.exp(2*x)
    return x,y,z

x,y,z = solve_system(f,g,0,3,0,1,20)


print("Решение, полученное алгоритмом")
print(x[-1],y[-1],z[-1])
print("Точное решение")
print(solution(1,3,0))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

def F(y,z):
    return -2*y+4*z
def G(y,z):
    return -y+3*z

Y, Z = np.meshgrid(y, z)

U = F(Y,Z)
V = G(Y,Z)

plt.figure()
plt.title('Фазовый портрет')
Q = plt.quiver(Y, Z, U, V, units='width')
qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
plt.show()
```
