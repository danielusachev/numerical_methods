```python
from scipy import interpolate
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
```

```python
#x0 - сетка
def cubic_interp1d(x0, x, y):

    x = np.asfarray(x)
    y = np.asfarray(y)


    # проверяем, отсортирован ли массив
    if np.any(np.diff(x) < 0):
        indexes = np.argsort(x)
        x = x[indexes]
        y = y[indexes]

    size = len(x)

    xdiff = np.diff(x)
    ydiff = np.diff(y)

    Li = np.empty(size)
    Li_1 = np.empty(size-1)
    z = np.empty(size)

    #решим[L][y] = [B]
    Li[0] = sqrt(2*xdiff[0])
    Li_1[0] = 0.0
    B0 = 0.0
    z[0] = B0 / Li[0]

    for i in range(1, size-1, 1):
        Li_1[i] = xdiff[i-1] / Li[i-1]
        Li[i] = sqrt(2*(xdiff[i-1]+xdiff[i]) - Li_1[i-1] * Li_1[i-1])
        Bi = 6*(ydiff[i]/xdiff[i] - ydiff[i-1]/xdiff[i-1])
        z[i] = (Bi - Li_1[i-1]*z[i-1])/Li[i]

    i = size - 1
    Li_1[i-1] = xdiff[-1] / Li[i-1]
    Li[i] = sqrt(2*xdiff[-1] - Li_1[i-1] * Li_1[i-1])
    Bi = 0.0
    z[i] = (Bi - Li_1[i-1]*z[i-1])/Li[i]

    # решим [L.T][x] = [y]
    i = size-1
    z[i] = z[i] / Li[i]
    for i in range(size-2, -1, -1):
        z[i] = (z[i] - Li_1[i-1]*z[i+1])/Li[i]

    index = x.searchsorted(x0)
    np.clip(index, 1, size-1, index)

    xi1, xi0 = x[index], x[index-1]
    yi1, yi0 = y[index], y[index-1]
    zi1, zi0 = z[index], z[index-1]
    hi1 = xi1 - xi0

    f0 = zi0/(6*hi1)*(xi1-x0)**3 + \
         zi1/(6*hi1)*(x0-xi0)**3 + \
         (yi1/hi1 - zi1*hi1/6)*(x0-xi0) + \
         (yi0/hi1 - zi0*hi1/6)*(xi1-x0)
    return f0
```

```python
def f_smooth(x):
    return (x**3)*np.sin(x)



def f_oscillating(x):
    return (np.sin(20*x))

def f_discontinuous(x):
    return 1/(x-10.0001)



x = np.linspace(0, 10, 31)
grid1 = np.linspace(0, 10, 11)
N = 31
grid2 = np.linspace(0, 10, N)

y = f_smooth(x)

plt.scatter(x, y)

plt.plot(grid1, cubic_interp1d(grid1, x, y))

plt.plot(grid2, cubic_interp1d(grid2, x, y))

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Гладкая функция и ее сплайн на двух сетках')
plt.show()


error = y[:N] - cubic_interp1d(grid2, x, y)
plt.plot(grid2, error)
plt.xlabel('X')
plt.ylabel('E(x)')
plt.title('График ошибки')
plt.show()




y = f_oscillating(x)

plt.scatter(x, y)

plt.plot(grid1, cubic_interp1d(grid1, x, y))

plt.plot(grid2, cubic_interp1d(grid2, x, y))

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Осциллирующая функция и ее сплайн на двух сетках')

plt.show()

error = y[:N] - cubic_interp1d(grid2, x, y)
plt.plot(grid2, error)
plt.xlabel('X')
plt.ylabel('E(x)')
plt.title('График ошибки')
plt.show()


y = f_discontinuous(x)

plt.scatter(x, y)

plt.plot(grid1, cubic_interp1d(grid1, x, y))

plt.plot(grid2, cubic_interp1d(grid2, x, y))

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Разрывная функция и ее сплайн на двух сетках')

plt.show()

error = y[:N] - cubic_interp1d(grid2, x, y)
plt.plot(grid2, error)
plt.xlabel('X')
plt.ylabel('E(x)')
plt.title('График ошибки')
plt.show()
```
