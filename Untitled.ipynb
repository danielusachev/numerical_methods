```python
from tkinter import *
from tkinter import messagebox
from tkinter import scrolledtext
from PIL import ImageTk,Image
import numpy as np
from tkinter import ttk
import pandas as pd
import matplotlib.pyplot as plt
import random
import os.path
import os
from tkinter import filedialog

class Parameters:
    a_st = 0
    a_zt = 0
    a_pw = 0
    b_pw = 0
    grid = 0
    @staticmethod
    def p_func(w):
        return Parameters.a_pw*w*(Parameters.b_pw-w)
    @staticmethod
    def s_func(t):
        return Parameters.a_st*t + np.sin(t)
    @staticmethod
    def z_func(t):
        return Parameters.a_zt*t + np.cos(t)
    @staticmethod
    def generategrid(a,b,n):
        Parameters.grid = np.linspace(a,b,n)

def tabulate(a,b,n):
    Parameters.generategrid(a,b,n)
    p_w = [Parameters.p_func(w) for w in Parameters.grid]
    s_t = [Parameters.s_func(t) for t in Parameters.grid]
    z_t = [Parameters.z_func(t) for t in Parameters.grid]

    df = pd.DataFrame(columns=["x", "p_w", "s_t", "z_t"])
    df["x"] = Parameters.grid
    df["p_w"] = p_w
    df["s_t"] = s_t
    df["z_t"] = z_t
    df.to_csv("tabulated.csv", index=False)
    print(df)
    
    
def draw_plots():
    p_w = [Parameters.p_func(w) for w in Parameters.grid]
    s_t = [Parameters.s_func(t) for t in Parameters.grid]
    z_t = [Parameters.z_func(t) for t in Parameters.grid]
    fig, ax = plt.subplots()
    ax.plot(Parameters.grid, p_w,label='p(w)')
    ax.plot(Parameters.grid, s_t,label='S(t)')
    ax.plot(Parameters.grid, z_t,label='z(t)')
    ax.set(xlabel='x', ylabel='y',
    title='S(t),z(t) and p(w)')
    ax.grid()
    ax.legend(loc='upper right')
    fig.savefig("plot.png")
    plt.show()

def interpolate():
    return random.random()
    
def button_grid_clicked():
    a_st = txt1.get()
    a_zt = txt2.get()
    a_pw = txt3.get()
    b_pw = txt4.get()
    if (not a_st.isdigit() or not a_zt.isdigit() or not a_pw.isdigit() or not b_pw.isdigit()):
        messagebox.showerror("Ошибка", "Необходимо ввести числа")
        return
    Parameters.a_st = float(a_st)
    Parameters.a_zt = float(a_zt)
    Parameters.a_pw = float(a_pw)
    Parameters.b_pw = float(b_pw)
    
    a = txt_grid_a.get()
    b = txt_grid_b.get()
    n = txt_grid_n.get()
    if (not a.isdigit() or not b.isdigit() or not n.isdigit()):
        messagebox.showerror("Ошибка", "Необходимо ввести числа")
        return

    tabulate(int(a), int(b), int(n))
    messagebox.showinfo("Готово", "Табулированные функции записаны в файл tabulated.csv")
    
def button2_clicked():
    
    window.filename =  filedialog.askopenfilename(initialdir = os.getcwd(),title = "Select file",filetypes = 
                                                (("csv files","*.csv"),("all files","*.*")))
    df = pd.read_csv(window.filename)
    print(df)
    
def button3_clicked():
    messagebox.showinfo("Готово", "Интеграл интерполирован")
def button_save_coefs():
    df = pd.DataFrame(columns=["c"])
    df["c"] = [interpolate()]
    df.to_csv("coefs.csv", index=False)
    print(df)
    messagebox.showinfo("Готово", "Коэффициенты сохранены в файл coefs.csv")

def button4_clicked():
    draw_plots()
    messagebox.showinfo("Готово", "График сохранен в файл plot.png")
     
def button5_clicked():
    pass

def on_closing():
     window.destroy()
        
window = Tk()
 
window.title("Проект")
 
window.geometry('600x450')

tab_control = ttk.Notebook(window)
 
tab1 = ttk.Frame(tab_control)
tab2 = ttk.Frame(tab_control)
tab3 = ttk.Frame(tab_control)

tab_control.add(tab1, text='Ввод данных') 
tab_control.add(tab2, text='Табулирование интеграла')
tab_control.add(tab3, text='Задача Коши')

lbl = Label(tab1, text="Ввод параметров функций")
lbl.grid(column=0, row=0)

lbl1 = Label(tab1,width=10, text="a для S(t):")
lbl1.grid(column=0, row=1)

txt1 = Entry(tab1,width=12)
txt1.grid(column=1, row=1, padx=5, pady=5)


lbl2 = Label(tab1,width=10, text="a для z(t):")
lbl2.grid(column=0, row=2)

txt2 = Entry(tab1,width=12)
txt2.grid(column=1, row=2, padx=5, pady=5)


lbl3 = Label(tab1,width=10, text="a для p(w):")
lbl3.grid(column=0, row=3)

txt3 = Entry(tab1,width=12)
txt3.grid(column=1, row=3, padx=5, pady=5)


lbl4 = Label(tab1,width=10, text="b для p(w):")
lbl4.grid(column=0, row=4)

txt4 = Entry(tab1,width=12)
txt4.grid(column=1, row=4, padx=5, pady=5)

lbl = Label(tab1, text="Ввод параметров сетки")
lbl.grid(column=0, row=6)

lbl = Label(tab1,width=10, text="a:")
lbl.grid(column=0, row=7)

txt_grid_a = Entry(tab1,width=12)
txt_grid_a.grid(column=1, row=7, padx=5, pady=5)

lbl = Label(tab1,width=10, text="b:")
lbl.grid(column=0, row=8)

txt_grid_b = Entry(tab1,width=12)
txt_grid_b.grid(column=1, row=8, padx=5, pady=5)

lbl = Label(tab1,width=10, text="n:")
lbl.grid(column=0, row=9)

txt_grid_n = Entry(tab1,width=12)
txt_grid_n.grid(column=1, row=9, padx=5, pady=5)

btn1 = Button(tab1,width=20, text="Табулировать функции", command=button_grid_clicked)
btn1.grid(column=1, row=10)


lbl = Label(tab2,width=20, text="Считать p(w) из файла:")
lbl.grid(column=0, row=0)

btn2 = Button(tab2,width=20, text="Выбрать файл ...", command=button2_clicked)
btn2.grid(column=0, row=1,padx=5,pady=5)
btn3 = Button(tab2,width=30, text="Интерполировать интеграл", command=button3_clicked)
btn3.grid(column=0, row=2,padx=5,pady=5)
btn4 = Button(tab2,width=30, text="Сохранить коэффициенты в файл", command=button_save_coefs)
btn4.grid(column=0, row=3,padx=5,pady=5)


lbl_3 = Label(tab3, text="Ввод параметров задачи Коши")
lbl_3.grid(column=0, row=0)

lbl1_3 = Label(tab3,width=10, text="X0:")
lbl1_3.grid(column=0, row=1)

txt1_3 = Entry(tab3,width=12)
txt1_3.grid(column=1, row=1, padx=5, pady=5)


lbl2_3 = Label(tab3,width=10, text="Y0:")
lbl2_3.grid(column=0, row=2)

txt2_3 = Entry(tab3,width=12)
txt2_3.grid(column=1, row=2, padx=5, pady=5)


lbl3_3 = Label(tab3,width=10, text="B:")
lbl3_3.grid(column=0, row=3)

txt3_3 = Entry(tab3,width=12)
txt3_3.grid(column=1, row=3, padx=5, pady=5)


lbl4_3 = Label(tab3,width=10, text="T:")
lbl4_3.grid(column=0, row=4)

txt4_3 = Entry(tab3,width=12)
txt4_3.grid(column=1, row=4, padx=5, pady=5)

btn4_3 = Button(tab3,width=20, text="Построить график", command=button4_clicked)
btn4_3.grid(column=0, row=5,padx=5,pady=5)

btn5_3 = Button(tab3,width=20, text="Решить задачу Коши", command=button5_clicked)
btn5_3.grid(column=0, row=6,padx=5,pady=5)

tab_control.pack(expand=1, fill='both')

window.protocol("WM_DELETE_WINDOW", on_closing)
window.mainloop()
```
