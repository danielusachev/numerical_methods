Используем метод трапеций

Простая формула
${\displaystyle
\int _{a}^{b}f(x)\,dx={\frac  {f(a)+f(b)}{2}}(b-a)+E(f),\qquad E(f)=-{\frac  {f''(\xi )}{12}}\left(b-a\right)^{3}}
$
<br>
Составная формула
${\displaystyle
\int _{a}^{b}f(x)\,dx\approx \sum _{{i=0}}^{{n-1}}{\frac  {f(x_{i})+f(x_{{i+1}})}{2}}(x_{{i+1}}-x_{{i}})}
$
<br>
Оценим погрешность через максимум второй производной
$
{\displaystyle \left|E(f)\right|\leqslant {\frac {\left(b-a\right)^{3}}{12n^{2}}}\max _{x\in [a,b]}\left|f''(x)\right|,{\frac {(b-a)^{3}}{12n^{2}}}={\frac {nh^{3}}{12}} = {\frac {h^{2}(b-a)}{12}}}$
<br>
Таким образом, формула трапеций имеет, второй порядок точности

```python
import scipy.integrate as integrate
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

class Parameters:
    @staticmethod
    def generategrid_even(a,b,n):
        return np.linspace(a,b,n)
    @staticmethod
    def generategrid_decreasing(a,b,n):
        grid = [0] * n
        grid[0] = a
        step = (b-a)/n
        for i in range(1,n):
            grid[i] = grid[i-1]+step
            step *= 0.999
        return grid

#метод трапеций
def num_integrate(func,grid):
    a = grid[0]
    b = grid[-1]
    dx = grid[1] - grid[0]
    Area = 0
    for i in range(1, n+1):
        x0 = a + (i-1)*dx
        x1 = a+i*dx
        Ai = dx * (func(x0) + func(x1))/ 2.
        Area = Area + Ai
        if i != n:
            dx = grid[i] - grid[i-1]
    return Area

def f_discontinuous(x):
    if x == 0:
        return 0
    return 1/x

def f_oscillating(x):
    if x == 0:
        return 1
    return (np.sin(20*x)/x)

A = 0
B = 10

ns = list(range(10,200))
errors = []
for n in ns:
    val1 = num_integrate(np.sqrt,Parameters.generategrid_even(A,B,n))
    val2 = integrate.quad(np.sqrt,A, B)[0]
    error = np.abs(val1-val2)
    errors.append(error)

plt.plot(ns, errors, 'ro')
plt.xlabel('Число разбиений N')
plt.ylabel('Модуль ошибки E(h)')
plt.title('График зависимости ошибки от числа разбиений для эксперимента 1 (гладкая функция)')
plt.show()



ns = list(range(10,200))
errors = []
for n in ns:
    val1 = num_integrate(f_oscillating,Parameters.generategrid_even(A,B,n))
    val2 = integrate.quad(f_oscillating,A, B)[0]
    error = np.abs(val1-val2)
    errors.append(error)
    
plt.plot(ns, errors, 'ro')
plt.xlabel('Число разбиений N')
plt.ylabel('Модуль ошибки E(h)')
plt.title('График зависимости ошибки от числа разбиений для эксперимента 2(осциллирующая функция)')
plt.show()


ns = list(range(10,200))
errors = []
for n in ns:
    val1 = num_integrate(f_discontinuous,Parameters.generategrid_even(A,B,n))
    val2 = integrate.quad(f_discontinuous,A, B)[0]
    error = np.abs(val1-val2)
    errors.append(error)
    
plt.plot(ns, errors, 'ro')
plt.xlabel('Число разбиений N')
plt.ylabel('Модуль ошибки E(h)')
plt.title('График зависимости ошибки от числа разбиений для эксперимента 3(разрывная функция)')
plt.show()
```
